## Importar trackings

```python
from datetime import date
from datetime import timedelta
from blip_report_requisitor import Requisitor

rows = []
rows_input = []

for index, key in key.items():    
    for day in range(0,15):
        start_date = date.today() - timedelta(days=day)
        
        # Set BOT Buider key
        req = Requisitor(key)
        
        reports = [req.getCustomReport(x, start_date, start_date) for x in t]
        
        start_date = start_date.strftime('%m/%d/%Y')
        for count, category in enumerate(reports):
            if type(category) is not int:                
                for action in category:
                    infos = action['acao'].split('|')
                    if len(infos) < 5:
                        for i in range(5 - len(infos)):
                            infos.append('')
                            
                    row = [index, trackings[count], infos[0], action['total'], start_date, infos[1], infos[2], infos[3], infos[4]]
                    
                    if trackings[count].find('_input') == -1 and trackings[count].find('_invalid') == -1:
                        rows.append(row)
                    else:
                        rows_input.append(row)
```

## Criar dataframe

```python
header = ['buider','category','action','amount','date','ies','campus','cpf','contact.identity']

import pandas as pd

data = pd.DataFrame(rows, columns = header)
data_input = pd.DataFrame(rows_input, columns = header)

display(data, data_input)
```
